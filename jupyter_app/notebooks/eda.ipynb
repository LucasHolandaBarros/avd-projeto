```python
# O que está basicamente fazendo?

# 1. Carrega o csv clean e interpreta as datas
# 2. Dá uma visão geral dos dados e tipos
# 3. Mostra estatísticas e missing values
# 4. Cria um índice date time
# 5. Plota gráficos básicos de série temporal, dispersão e histograma

FILE_PATH = r"C:\Users\euque\OneDrive\Documentos\Cesar School\5° Período - CC Cesar School\Análise e Visualização de Dados\avd-projeto\data\processed\concatenado_clean_20251125_202250.csv"  

import pandas as pd
import matplotlib.pyplot as plt

# Load
df = pd.read_csv(FILE_PATH, parse_dates=['DATA'], dayfirst=True, low_memory=False)

# Quick overview
print('Rows, Columns:', df.shape)
display(df.head())
display(df.dtypes)

# Basic stats for numeric columns
display(df.describe())

# Missing values summary
missing = df.isnull().sum().sort_values(ascending=False)
display(missing[missing>0])

# Convert DATETIME for timeseries if HORA_UTC exists
if 'HORA_UTC' in df.columns:
    try:
        df['HORA_UTC'] = df['HORA_UTC'].astype(str).str.zfill(2)
        df['DATETIME'] = pd.to_datetime(df['DATA'].dt.strftime('%Y-%m-%d') + ' ' + df['HORA_UTC'] + ':00:00')
        df = df.set_index('DATETIME')
        display(df.head())
    except Exception as e:
        print('Não foi possível montar DATETIME automaticamente:', e)

# Plot examples (one plot per cell as recommended)
# 1) Time series of UMIDADE_REL (if present)
if 'UMIDADE_REL' in df.columns:
    plt.figure(figsize=(12,4))
    df['UMIDADE_REL'].dropna().iloc[:1000].plot(title='UMIDADE_REL (sample)')
    plt.ylabel('Umidade (%)')
    plt.show()

# 2) Scatter: TEMPERATURA_BULBO_SECO vs UMIDADE_REL
if 'TEMPERATURA_BULBO_SECO' in df.columns and 'UMIDADE_REL' in df.columns:
    subset = df[['TEMPERATURA_BULBO_SECO', 'UMIDADE_REL']].dropna()
    plt.figure(figsize=(6,4))
    plt.scatter(subset['TEMPERATURA_BULBO_SECO'], subset['UMIDADE_REL'])
    plt.xlabel('TEMPERATURA_BULBO_SECO')
    plt.ylabel('UMIDADE_REL')
    plt.title('Temperatura x Umidade (scatter)')
    plt.show()


# 3) Histogram of TEMPERATURA_BULBO_SECO
if 'TEMPERATURA_BULBO_SECO' in df.columns:
    plt.figure(figsize=(6,4))
    df['TEMPERATURA_BULBO_SECO'].dropna().hist(bins=30)
    plt.title('Distribuição da TEMPERATURA_BULBO_SECO')
    plt.show()

print('EDA concluída — ajuste FILE_PATH se necessário.')
```
